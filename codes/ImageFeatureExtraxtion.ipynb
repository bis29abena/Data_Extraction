> This is the code for reading images using the opencv library

`cv2.imread('image_path')`



```python
imagePath = "../data/dog.jpg"

img = cv2.imread(imagePath)
```

> Use matplotlib to plot your image and show it on the screen

```python
import matplotlib.pyplot as plt
import matplotlib

plt.imshow(img)
```

> This is a colored image so therefore the image will be a 3D matrix
> We can see the that using the `img.shape`

```python
img.shape
```

> The above output shows that the image is 200 x 300 (height, width) pixel image and the 3 represents the number channels meaning is a coloured image

```python
img
```

> The output represents each pixel in the image whic is in an RGB form so to see the output of the first pixel we use `img[0][0]`

```python
img[0][0]
```

> We can also resize the image to a 32 X 32 size and also show it using a matplotlib
>
> We resize the image without maintaining the aspect ratio or without using any intertpolation function

```python
size = (32, 32)
    
resized_image = cv2.resize(img, size)

plt.imshow(resized_image)
```

```python
resized_image.shape
```

> Now after reszing the image without keeping the aspect ratio the image has been distorted
>
> And with this you can't really see the what is actually in that image 

#### Now let's resize the image whiles maintaing aspect ratio

```python
r = 32.0 / img.shape[0]

dim = (int(img.shape[1] * r), 32)

resized_image = cv2.resize(img, dim, interpolation=cv2.INTER_AREA)

plt.imshow(resized_image)
```

```python
resized_image.shape
```

### We can convert an image from colored to grayscale

```python
grayscale_image = cv2.imread(imagePath, cv2.IMREAD_GRAYSCALE)

plt.imshow(grayscale_image)
```

```python
grayscale_image.shape
```

> Now in the above output since the image is in grayscale the dimension of the shape is in 2D 

```python
grayscale_image
```

> since the image has been converted into grayscale there is only a single intensity value in the pixels

### We can change the dimension of the grayscale image to show how it really looks like in a 2D dimension

```python
import numpy as np

expanded_image = np.expand_dims(grayscale_image, axis=2)
expanded_image
```
