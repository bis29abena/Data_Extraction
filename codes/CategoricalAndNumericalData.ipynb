```python
import pandas as pd
%matplotlib inline
```

```python
exams_data = pd.read_csv("../data/exams.csv", quotechar= '"')
exams_data
```

## Standardizing Data for the Numerical fields.

## We can find the means for each numerical column

```python
math_score_average = exams_data["math score"].mean()
reading_score_average = exams_data["reading score"].mean()
writing_score_average = exams_data["writing score"].mean()

print(f"Math score average: {math_score_average}")
print(f"Reading score average: {reading_score_average}")
print(f"Writing score average: {writing_score_average}")
```

### we don't know the percentage of each score so standardizing those columns can help our model in trainging

```python
from sklearn import preprocessing

exams_data[["math score"]] = preprocessing.scale(exams_data[["math score"]])
exams_data[["reading score"]] = preprocessing.scale(exams_data[["reading score"]])
exams_data[["writing score"]] = preprocessing.scale(exams_data[["writing score"]])
```

```python
exams_data
```

```python
math_score_average = exams_data["math score"].mean()
reading_score_average = exams_data["reading score"].mean()
writing_score_average = exams_data["writing score"].mean()

print(f"Math score average: {math_score_average}")
print(f"Reading score average: {reading_score_average}")
print(f"Writing score average: {writing_score_average}")
```
